```python
class mydata(Dataset):
    def __init__(self,root_path,label_path):
        self.root_path=root_path
        self.label_path=label_path
        self.path= os.path.join(self.root_path,self.label_path)
        self.list_path= os.listdir(self.path)



    def __getitem__(self, idx):
        image_name =self.list_path[idx]
        image_item_path = os.path.join(self.root_path,self.label_path,image_name)
        image = Image.open(image_item_path)
        label = self.label_path
        return image,label
    def __len__(self):
        return len(self.list_path)
```

```python
#蚂蚁数据集
root_path=r"C:\Users\66\PycharmProjects\pytorch\data_set\train"
ants_lable_path="ants"
ants_data = mydata(root_path,ants_lable_path)

#蜜蜂数据集
bee_lable_path="bees"
bees_data = mydata(root_path,bee_lable_path)
bees_data[0]
```

```python
merge_data= ants_data+bees_data
```
